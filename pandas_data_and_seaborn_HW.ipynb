```python
import pandas as pd

iris_df = pd.read_csv('/Users/heyodogo/Downloads/Iris.csv')

iris_df.drop(columns=['Id'], inplace=True)

iris_df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

print(iris_df)
```

```python
summary_statistics = iris_df.describe().loc[['mean', '50%', 'std']]

summary_file_path = '/Users/heyodogo/Downloads/Iris_summary.csv'

summary_statistics.to_csv(summary_file_path)

print(summary_statistics)
```

```python
import seaborn as sns

sns.set_style('whitegrid')

sns.pairplot(iris_df, hue='species')

numerical_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for i in numerical_columns:
    sns.boxplot(x='species', y=i, data=iris_df)
```

```python
def visualize_iris(data, plot_type="box"):
    numerical_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
    
    if plot_type == "strip":
        for i in numerical_columns:
            sns.stripplot(x='species', y=i, data=data)
    
    elif plot_type == "violin":
        for i in numerical_columns:
            sns.violinplot(x='species', y=i, data=data)
    
    elif plot_type == "box":
        for i in numerical_columns:
            sns.boxplot(x='species', y=i, data=data)
            
    else:
        print('aint right dawg')
```

```python
visualize_iris(iris_df, plot_type="box")
```

```python
visualize_iris(iris_df, plot_type="violin")
```

```python
visualize_iris(iris_df, plot_type="strip")
```

```python
visualize_iris(iris_df, plot_type='wow')
```
